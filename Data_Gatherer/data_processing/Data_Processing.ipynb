```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
import pandas as pd
```

```python
# This will be the module to retrieve our data and place it in a data frame.

csv_2009 = pd.read_csv("2009.csv")
csv_2010 = pd.read_csv("2010.csv")
csv_2011 = pd.read_csv("2011.csv")
csv_2012 = pd.read_csv("2012.csv")
csv_2013 = pd.read_csv("2013.csv")
csv_2014 = pd.read_csv("2014.csv")
csv_2015 = pd.read_csv("2015.csv")
csv_2016 = pd.read_csv("2016.csv")
csv_2017 = pd.read_csv("2017.csv")
csv_2018 = pd.read_csv("2018.csv")

csv_capacity = pd.read_csv("Park_Capacity_Year.csv")


df_2009 = pd.DataFrame(csv_2009)
df_2010 = pd.DataFrame(csv_2010)
df_2011 = pd.DataFrame(csv_2011)
df_2012 = pd.DataFrame(csv_2012)
df_2013 = pd.DataFrame(csv_2013)
df_2014 = pd.DataFrame(csv_2014)
df_2015 = pd.DataFrame(csv_2015)
df_2016 = pd.DataFrame(csv_2016)
df_2017 = pd.DataFrame(csv_2017)
df_2018 = pd.DataFrame(csv_2018)
df_capacity = pd.DataFrame(csv_capacity)
```

```python
#csv_2009.groupby("Tm").count()


#df_capacity["Difference %"] = (100 / df_capacity["Capacity D"]) * df_capacity["Capacity N"]
#df_capacity.head(20)
df_capacity_2009 = df_capacity[df_capacity["Year"]==2009]
df_capacity_2009.head()

df_capacity_2010 = df_capacity[df_capacity["Year"]==2010]
df_capacity_2010.head()

df_capacity_2011 = df_capacity[df_capacity["Year"]==2011]
df_capacity_2011.head()

df_capacity_2012 = df_capacity[df_capacity["Year"]==2012]
df_capacity_2012.head()

df_capacity_2013 = df_capacity[df_capacity["Year"]==2013]
df_capacity_2013.head()

df_capacity_2014 = df_capacity[df_capacity["Year"]==2014]
df_capacity_2014.head()

df_capacity_2015 = df_capacity[df_capacity["Year"]==2015]
df_capacity_2015.head()

df_capacity_2016 = df_capacity[df_capacity["Year"]==2016]
df_capacity_2016.head()

df_capacity_2017 = df_capacity[df_capacity["Year"]==2017]
df_capacity_2017.head()

df_capacity_2018 = df_capacity[df_capacity["Year"]==2018]
df_capacity_2018.head()
```

```python
csv_2009 = csv_2009.rename(columns={"Tm": "Team"})
csv_2010 = csv_2010.rename(columns={"Tm": "Team"})
csv_2011 = csv_2011.rename(columns={"Tm": "Team"})
csv_2012 = csv_2012.rename(columns={"Tm": "Team"})
csv_2013 = csv_2013.rename(columns={"Tm": "Team"})
csv_2014 = csv_2014.rename(columns={"Tm": "Team"})
csv_2015 = csv_2015.rename(columns={"Tm": "Team"})
csv_2016 = csv_2016.rename(columns={"Tm": "Team"})
csv_2017 = csv_2017.rename(columns={"Tm": "Team"})
csv_2018 = csv_2018.rename(columns={"Tm": "Team"})

csv_2009.count()
csv_2018.head(10)
```

```python
#inner join in python pandas
merge_2009 = pd.merge(df_capacity_2009, csv_2009, on='Team', how='inner')
merge_2010 = pd.merge(df_capacity_2010, csv_2010, on='Team', how='inner')
merge_2011 = pd.merge(df_capacity_2011, csv_2011, on='Team', how='inner')
merge_2012 = pd.merge(df_capacity_2012, csv_2012, on='Team', how='inner')
merge_2013 = pd.merge(df_capacity_2013, csv_2013, on='Team', how='inner')
merge_2014 = pd.merge(df_capacity_2014, csv_2014, on='Team', how='inner')
merge_2015 = pd.merge(df_capacity_2015, csv_2015, on='Team', how='inner')
merge_2016 = pd.merge(df_capacity_2016, csv_2016, on='Team', how='inner')
merge_2017 = pd.merge(df_capacity_2017, csv_2017, on='Team', how='inner')
merge_2018 = pd.merge(df_capacity_2018, csv_2018, on='Team', how='inner')

#merge_2009.count()
merge_2011.head(200)
```

```python
games = [merge_2009, merge_2010, merge_2011, merge_2012, merge_2013, merge_2014, merge_2015, merge_2016, merge_2017, merge_2018]
concat_games = pd.concat(games,ignore_index=True)
```

```python

concat_games.count()
concat_games["Capacity"] = "0"
concat_games.head()

#concat_games.iloc[1, 5]




concat_games = concat_games.reset_index()
#concat_games.columns[0] = 'New_ID'
#concat_games["New_ID"] = concat_games.index


concat_games.head(10)
```

```python
concat_games["Capacity"] = concat_games["index"].apply(lambda x: concat_games.iloc[x, 3] if concat_games.iloc[x, 6]=='D' else concat_games.iloc[x, 4])
```

```python
concat_games = concat_games.drop(['Capacity D', 'Capacity N'], axis=1)

concat_games["Attendance Rate"] = (100/concat_games["Capacity"]) * concat_games["Attendance"]


```

```python
#Add weather Data
Boston_DayGames = pd.read_csv("Boston_DayGames.csv")
Boston_NightGames = pd.read_csv("Boston_NightGames.csv")
Chicago_DayGames = pd.read_csv("Chicago_DayGames.csv")
Chicago_NightGames = pd.read_csv("Chicago_NightGames.csv")
```

```python
df_Boston_DayGames = pd.DataFrame(Boston_DayGames)
df_Boston_DayGames["D/N"] = "D"
df_Boston_DayGames["Team"] = "BOS"
df_Boston_DayGames = df_Boston_DayGames.rename(columns={"Unnamed: 0": "API DATE Format"})

df_Boston_NightGames = pd.DataFrame(Boston_NightGames)
df_Boston_NightGames["D/N"] = "N"
df_Boston_NightGames["Team"] = "BOS"
df_Boston_NightGames = df_Boston_NightGames.rename(columns={"Unnamed: 0": "API DATE Format"})



df_Chicago_DayGames = pd.DataFrame(Chicago_DayGames)
df_Chicago_DayGames["D/N"] = "D"
df_Chicago_DayGames["Team"] = "CHC"
df_Chicago_DayGames = df_Chicago_DayGames.rename(columns={"Unnamed: 0": "API DATE Format"})


df_Chicago_NightGames = pd.DataFrame(Chicago_NightGames)
df_Chicago_NightGames["D/N"] = "N"
df_Chicago_NightGames["Team"] = "CHC"
df_Chicago_NightGames = df_Chicago_NightGames.rename(columns={"Unnamed: 0": "API DATE Format"})

```

```python

df_Boston_DayGames.head()

concat_games_1 = pd.merge(concat_games, df_Boston_DayGames, on=['API DATE Format','Team','D/N'], how='inner')
concat_games_2 = pd.merge(concat_games, df_Boston_NightGames, on=['API DATE Format','Team','D/N'], how='inner')
concat_games_3 = pd.merge(concat_games, df_Chicago_DayGames, on=['API DATE Format','Team','D/N'], how='inner')
concat_games_4 = pd.merge(concat_games, df_Chicago_NightGames, on=['API DATE Format','Team','D/N'], how='left')

games = [concat_games_1, concat_games_2, concat_games_3, concat_games_4]
concat_games_new = pd.concat(games,ignore_index=True)

concat_games.head(10)

concat_games_new.to_csv("merge_sample_before_filter.csv")

#concat_games_1.count()
#concat_games[(concat_games['D/N']=='D') & (concat_games['Team']=='BOS')].count()
```

```python
chicago_attendence_rate_mean = concat_games_new["Attendance Rate"][concat_games_new["Team"]=="CHC"].mean()
boston_attendence_rate_mean = concat_games_new["Attendance Rate"][concat_games_new["Team"]=="BOS"].mean()
```

```python
print(chicago_attendence_rate_mean)
print(boston_attendence_rate_mean)
```

```python
stats.ttest_ind(concat_games_new["Attendance Rate"][concat_games_new["Team"]=="CHC"], concat_games_new["Attendance Rate"][concat_games_new["Team"]=="BOS"], equal_var=False)
```

```python
#H0 = There is no difference between fan games attendance for Chicago Cubs and Boston Red Sox
#(Mean attendance Cubs = Mean Attendance Red Sox)

#Ha = Mean attendance Cubs <> Mean Attendance Red Sox
#p = 0.05

#T test P value = 1.62733532339865e-90
#< 0.05 So we reject H0 Hipotesis

#So we adopt Ha
#Mean attendance Cubs <> Mean Attendance Red Sox
#with a P Value of 0.05

```

```python
# Now we are going to filter out those days with temperatures higher than 60 unless is raining more than 0.5 inch

concat_games_new_filtered = concat_games_new[(concat_games_new['Temperature']<60)|(concat_games_new['Precipitation']>=0.09)]

concat_games_new_filtered.head(1000)
concat_games_new_filtered.to_csv("merge_sample_after_filter.csv")
```
