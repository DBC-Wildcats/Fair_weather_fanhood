```python
#NWU Data Science Boot Camp
#Project 1
#Weather data API script request


#call modules and dependencies
import pandas as pd
import requests
import os
import glob
from config import apiKey, CHICAGO, BOSTON, WRIGLEY, FENWAY
from pprint import pprint
```

```python
os.chdir("./API CSVs")
keyCounter = 1000
```

```python
keyIndex = 2
baseURL = f"https://api.darksky.net/forecast/{apiKey[keyIndex]}/"
#need to append {city},{time}" string to baseURL in get
print(baseURL)
```

```python
#test api request format
query = str(baseURL+f'{WRIGLEY}'+","+"2009-04-01T13:00:00")
response = requests.get(query).json()
pprint(response)
print(f"CHI:")
#pprint([response['currently']['icon'],response['currently']['summary'],response['currently']['temperature'],
#       response['currently']['precipIntensity']])
```

```python
#Dictionaries for requests
chiDay = {}
chiNight = {}
bosDay = {}
bosNight =  {}
```

```python
#CSVs and dataframe
fileList = ["2009.csv", "2010.csv", "2011.csv", "2012.csv", "2013.csv",
            "2014.csv","2015.csv", "2016.csv", "2017.csv", "2018.csv"]
apiList = []

#find all csv files in the folder
#use glob pattern matching -> extension = 'csv'
#save result in list -> all_filenames
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
print(all_filenames)

#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')

#credited:
#https://stackoverflow.com/questions/9234560/find-all-csv-files-in-a-directory-using-python/12280052
```

```python
#cleaning up the DF
apiRequestDF = pd.DataFrame(combined_csv)
apiRequestDF = apiRequestDF.drop(['GameNo'], axis = 1)
apiRequestDF = apiRequestDF.rename(columns = {'Tm':'Team', 'API DATE Format':'Date'})
print(f"HEAD--")
print(apiRequestDF.head(4))
print(f"TAIL--")
print(apiRequestDF.tail(4))
print(apiRequestDF.count())
print(apiRequestDF.axes)
print(f"_____________________________________")
print(f"_____________________________________")
```

```python
  #request as per each row in the current in memory dataframe
for index, row in apiRequestDF.iterrows():
    #chicago day game request
    if (row["Team"] == "CHC" and row["D/N"] == 'D'):
        keyIndex = 4
        queryWrigleyD = baseURL+f"{WRIGLEY},"+f'{row["Date"]}'+f"T13:00:00"
        keyCounter -= 1
        responseWD = requests.get(queryWrigleyD).json()
        chiDay.update({row["Date"]:{"Icon":responseWD["currently"]["icon"],
                                    "Precipitation":responseWD["currently"]["precipIntensity"],
                                    "Temperature":responseWD["currently"]["temperature"],
                                    "Summary":responseWD["currently"]["summary"]}})
        print(f'Request:  {queryWrigleyD} complete')
print("\n------------------------------------")
print("ALL REQUESTS COMPLETE")
print("--------------------------------------\n")        
print(f'Remaining Key gets:  {keyCounter}')
```

```python
#boston day game request
findCt  = 0
for index, row in apiRequestDF.iterrows():
    if (row["Team"] == "BOS" and row["D/N"] == 'D'):
        findCt = findCt + 1
        keyIndex = 4    
#print(f'total finds:  {findCt}')
#    if (findCt <= keyCounter):
#        print(f"you have {keyCOunter - findCt} remaining ")
#    else:
#        print("NOT ENOUGH, SWITCH KEYS!")
        queryFenwayD = baseURL+f"{FENWAY},"+f'{row["Date"]}'+f"T13:00:00"
        keyCounter -= 1
        responseFD= requests.get(queryFenwayD).json()
        bosDay.update({row["Date"]:{"Icon":responseFD["currently"]["icon"],
                                    "Precipitation":responseFD["currently"]["precipIntensity"],
                                    "Temperature":responseFD["currently"]["temperature"],
                                    "Summary":responseFD["currently"]["summary"]}})
        print(f'Request:  {queryFenwayD} complete')
print("\n------------------------------------")
print("ALL REQUESTS COMPLETE")
print("--------------------------------------\n")        
print(f'Remaining Key gets:  {keyCounter}')      
print(f'total finds: {findCt}')
```
