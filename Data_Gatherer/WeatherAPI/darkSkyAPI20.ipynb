```python
#Dictionaries for requests
bosDayDict = {}
bosNightDict =  {}
chiDayDict = {}
chiNightDict = {}
```

```python
#boston day game weather request
#findCt  = 0
keyIndex = 0 
for index, row in bosDayDF.iterrows():
#    if (row["Team"] == "BOS" and row["D/N"] == 'D'):
#        findCt = findCt + 1           
#print(f'total finds:  {findCt}')
#    if (findCt <= keyCounter):
#        print(f"you have {keyCOunter - findCt} remaining ")
#    else:
#        print("NOT ENOUGH, SWITCH KEYS!")
        queryFenwayD = baseURL+f"{FENWAY},"+f'{row["Date"]}'+f"T13:00:00"
        keyCounter -= 1
        responseFD= requests.get(queryFenwayD).json()
        bosDayDict.update({row["Date"]:{"Icon":responseFD["currently"]["icon"],
                                    "Precipitation":responseFD["currently"]["precipIntensity"],
                                    "Temperature":responseFD["currently"]["temperature"],
                                    "Summary":responseFD["currently"]["summary"]}})
        print(f'Request:\n {queryFenwayD} complete')
print("\n------------------------------------")
print("ALL REQUESTS COMPLETE")
print(f"run:  {currentDT}")
print("--------------------------------------\n")        
print(f'Remaining Key gets:  {keyCounter}')      
#print(f'total finds: {findCt}')
```

```python
#boston night game weather request
reqCt = 0
keyIndex = 0
for index, row in bosNightDF.iterrows():
#    if (row["Team"] == "BOS" and row["D/N"] == 'N'):
        reqCt += 1
        
#print(f'total finds:  {findCt}')
#if (findCt <= keyCounter):
#    print(f"you have {keyCounter - findCt} remaining ")
#else:
#    print("NOT ENOUGH, SWITCH KEYS!")              
        queryFenwayN = baseURL+f'{FENWAY}'+","+f'{row["Date"]}'+"T18:00:00"
        keyCounter -= 1
        responseFN = requests.get(queryFenwayN).json()
        bosNightDict.update({row["Date"]:{"Icon":responseFN["currently"]["icon"],
                                   "Precipitation":responseFN["currently"]["precipIntensity"],
                                   "Temperature":responseFN["currently"]["temperature"],
                                   "Summary":responseFN["currently"]["summary"]}})
        print(f'Request {reqCt}:\n  {queryFenwayN} complete')
print("\n------------------------------------")
print("ALL REQUESTS COMPLETE")
print(currentDT)
print("--------------------------------------\n")
print(f'Remaining Key gets:  {keyCounter}')      
print(f'total finds: {reqCt}')
```
